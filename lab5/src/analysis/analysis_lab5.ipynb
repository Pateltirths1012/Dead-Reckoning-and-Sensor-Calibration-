# Analysis for Circle data

- A plot showing the magnetometer data before and after the correction in your report.

```python
magnetic_field_circle = data_reader_circle_imu_csv[['mag_field.magnetic_field.x','mag_field.magnetic_field.y','mag_field.magnetic_field.z']]
magnetic_field_circle_df = pd.DataFrame(magnetic_field_circle)
magnetic_field_circle_df.head()
```

```python
from scipy.optimize import least_squares
from sklearn.linear_model import LinearRegression

x_meas_circle = magnetic_field_circle_df['mag_field.magnetic_field.x']
y_meas_circle = magnetic_field_circle_df['mag_field.magnetic_field.y']
z_meas_circle = magnetic_field_circle_df['mag_field.magnetic_field.z']
X_meas = np.array([x_meas_circle, y_meas_circle, z_meas_circle])


#completely made up data set for "perfect" magnetometer readings
field_strength = 20509e-9
angle = np.linspace(-np.pi, np.pi, 1771)
y_mag = field_strength * np.cos(angle) 
x_mag = field_strength * np.sin(angle) 
X_mag = np.array([x_mag, y_mag, np.zeros_like(angle)])

def distortion_model(X_meas, dist_params):
    x = dist_params[0] * (X_meas[0] - dist_params[4]) + dist_params[1]*(X_meas[1] - dist_params[5])
    y = dist_params[2] * (X_meas[0] - dist_params[4]) + dist_params[3]*(X_meas[1] - dist_params[5])
    z = dist_params[4] * (X_meas[2] - dist_params[8])
    X = np.array([x,y,z])
    return X

#this function finds the difference between a circle and my transformed measurement
def residual(p, X_mag, X_meas):
    res = (X_mag - distortion_model(X_meas, p)).flatten()
    # print(res)
    return res

#Initial guess for ellipsoid parameters
initial_guess = [0, 0, 0, 0, 0, 0, 0, 0, 0]

#Fitting the ellipsoid using least squares
result = least_squares(residual, initial_guess, args=(X_mag, X_meas))
params = result.x

def calibrate_magnetometer(X_mag, X_meas):
    # Compute hard iron bias
    hard_iron_bias = np.mean(X_meas, axis=1)
    
    # Compute soft iron matrix using linear regression
    reg = LinearRegression().fit(X_mag.T, X_meas.T)
    soft_iron_matrix = reg.coef_.T
    
    return hard_iron_bias, soft_iron_matrix

def remove_distortions(X_meas, hard_iron_bias, soft_iron_matrix):
    # Remove hard iron bias
    X_meas_corrected = X_meas - hard_iron_bias[:, np.newaxis]
    
    # Apply soft iron correction
    X_meas_corrected = np.linalg.pinv(soft_iron_matrix) @ X_meas_corrected
    
    return X_meas_corrected

# Calibrate magnetometer
hard_iron_bias, soft_iron_matrix = calibrate_magnetometer(X_mag, X_meas)

# Correct measurements
X_meas_corrected = remove_distortions(X_meas, hard_iron_bias, soft_iron_matrix)

X_model = distortion_model(X_meas_corrected, params)

plt.style.use("seaborn-v0_8-dark")

plt.figure(figsize=(8,6))
plt.plot(X_mag[1] + 0.000062, X_mag[0],label='No distortion')
plt.scatter(X_model[1] + 0.000062, X_model[0] , label='calibrated data')
plt.scatter(y_meas_circle, x_meas_circle, label='Measured Data')
plt.xlabel('E Component (Tesla)')
plt.ylabel('N Component (Tesla)')
plt.title('Magnetic Field (N vs E Components) Circle Data')
plt.grid(True)
plt.axis('equal')
plt.legend()
plt.savefig('circle_mag_field_nvsecomp.png')
plt.show()
```

- Low pass filter of magnetometer data, high pass filter of gyro data, complementary filter output, and IMU heading estimate as 4 subplots on one plot

- First lets do the low pass filter on magnetometer data

```python
from scipy.signal import butter, sosfilt, freqz

def butter_filter(raw_data, cutoff_freq, sample_freq, filter_type, filter_order):
    nyq_freq = sample_freq / 2
    sos = butter(N = filter_order, Wn = cutoff_freq / nyq_freq, btype = filter_type, analog = False, output='sos')
    filtered_data = sosfilt(sos, raw_data)
    return sos, filtered_data

#setting filter requirements
order = 1
sample_freq = 40 #because we collected our data with 40 hz writing to the register the same
cutoff_freq = 1

sos, y = butter_filter(yaw_calibrated_driving, cutoff_freq, sample_freq, "lowpass", 5)

plt.figure(figsize=(8,6))
plt.plot(time_driving, yaw_calibrated_driving, label='Calibrated yaw Magnetometer', color='y')
plt.plot(time_driving, y, label='filtered yaw Magnetometer')
plt.xlabel('Time (s)')
plt.ylabel('yaw (Tesla)')
plt.title("Low pass filter Magnetometer yaw")
plt.legend()
plt.grid(True)
plt.show()
```
